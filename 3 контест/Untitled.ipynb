```python
def readLangSet(file):
    return set((file.readline().strip() for _ in range(int(file.readline()))))

def apply():
    file = open('input.txt')
    
    N = int(file.readline())

    all_lang = readLangSet(file)
    known_lang = all_lang.copy()
    for _ in range(N - 1):
        langSet = readLangSet(file)
        all_lang.update(langSet)
        known_lang.intersection_update(langSet)

    """print(len(known_lang))
    for lang in known_lang:
        print(lang)
    print(len(all_lang))
    for lang in all_lang:
        print(lang)"""
```
